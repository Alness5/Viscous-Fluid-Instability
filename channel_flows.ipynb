The code below is for Channel Flows:

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func, y_range, title, A):
    y = np.linspace(y_range[0], y_range[1], 400)
    x = func(y)
    plt.plot(x, y)
    y_arrows = np.linspace(y_range[0], y_range[1], 9)  # Adjust number of arrows
    for y_val in y_arrows:
        x_val = func(y_val)
        plt.annotate('', xy=(x_val, y_val), xytext=(0, y_val),
                     arrowprops=dict(arrowstyle="->", color='red'))
    plt.plot([-8, 8], [1.05, 1.05], linewidth=5, color='black')
    plt.plot([-8, 8], [-1.05, -1.05], linewidth=5, color='black')
    plt.plot([0, 0], [-2, 2], 'k--')
    plt.ylim(-2, 2)
    plt.xlim(-4, 4)
    plt.xlabel('U(y)')
    plt.ylabel('y')
    plt.savefig(f'{A}.png', format='png', dpi=1200)
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_function(lambda x: 0*x, (-1, 1), title='No Flow', A='No Flow')
```

```python
plot_function(lambda y: 1-y**2, (-1, 1), title='Plane Poiseuille Flow', A= 'Plane Poiseuille Flow')
```

```python
plot_function(lambda y: y, (-1, 1), title='Plane Couette Flow', A='Plane Couette Flow')
```

```python
R=1
G=-4
plot_function(lambda y: (R*G/2)*(y**2-1) + y, (-1, 1), title='Plane Poiseuille + Couette Flow', A='Plane Poiseuille + Couette Flow')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_function2(func, y_range, title, A):
    y = np.linspace(y_range[0], y_range[1], 400)
    x = func(y)
    plt.plot(x, y)
    y_arrows = np.linspace(y_range[0], y_range[1], 9)
    for y_val in y_arrows:
        x_val = func(y_val)
        plt.annotate('', xy=(x_val, y_val), xytext=(0, y_val),
                     arrowprops=dict(arrowstyle="->", color='red'))
    plt.plot([-4, 4], [1.05, 1.05], linewidth=5, color='black')
    plt.plot([-4, 4], [-1.05, -1.05], linewidth=5, color='black')
    plt.plot([0, 0], [-1, 1], 'k--')
    arrow_xs = np.linspace(-3, 3, 7)
    for x0 in arrow_xs:
        plt.annotate('', xy=(x0, -1.5), xytext=(x0, -1.05),
                     arrowprops=dict(arrowstyle="->", color='blue', lw=1))
        plt.annotate('', xy=(x0, 1.05), xytext=(x0, 1.5),
                     arrowprops=dict(arrowstyle="->", color='blue', lw=1))
    plt.ylim(-2, 2)
    plt.xlim(-4, 4)
    plt.xlabel('U(y)')
    plt.ylabel('y')
    plt.title(title)
    plt.grid(True)
    plt.savefig(f'{A}.png', format='png', dpi=1200)
    plt.show()
```

```python
R = 1
V = -4
plot_function2(
    lambda y: (1 / (1 - np.cosh(R * V))) * (np.exp(R * V * y) - np.cosh(R * V) - np.sinh(R * V) * y),
    (-1, 1),
    title='Asymptotic Suction Profile (non-zero pressure gradient & V<0)',
    A='Asymptotic Suction Profile (p != 0)')
```

```python
V= -2

plot_function2(
    lambda y: (1 / (2 * np.sinh(R * V))) * (np.exp(R * V * y) - np.exp(-R * V)),
    (-1, 1),
    title='Asymptotic Suction Profile (zero pressure gradient & V<0)',
    A='Asymptotic Suction Profile (p = 0)'
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_unbounded_function(func, y_range, title, A):
    y = np.linspace(y_range[0], y_range[1], 400)
    x = func(y)
    plt.plot(x, y)
    y_arrows = np.linspace(y_range[0], y_range[1], 9)
    for y_val in y_arrows:
        x_val = func(y_val)
        plt.annotate('', xy=(x_val, y_val), xytext=(0, y_val),
                     arrowprops=dict(arrowstyle="->", color='red'))
    arrow_xs = np.linspace(-3, 3, 7)
    for x0 in arrow_xs:
        plt.annotate('', xy=(x0, -1), xytext=(x0, 0),
                     arrowprops=dict(arrowstyle="->", color='blue', lw=1))
    plt.plot([-4, 4], [0, 0], linewidth=5, color='black')
    plt.plot([0, 0], [-1, 4], 'k--')
    plt.xlim(-4, 4)
    plt.xlabel('U(y)')
    plt.ylabel('y')
    plt.title(title)
    plt.grid(True)
    plt.savefig(f'{A}.png', format='png', dpi=1200)
    plt.show()
```
