The code below is for Channel Flows:

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func, y_range, title):
    y = np.linspace(y_range[0], y_range[1], 400)
    x = func(y)
    plt.plot(x, y)
    y_arrows = np.linspace(y_range[0], y_range[1], 9)  # Adjust number of arrows
    for y_val in y_arrows:
        x_val = func(y_val)
        plt.annotate('', xy=(x_val, y_val), xytext=(0, y_val),
                     arrowprops=dict(arrowstyle="->", color='red'))
    plt.plot([-8, 8], [2.05, 2.05], linewidth=5, color='black')
    plt.plot([-8, 8], [-2.05, -2.05], linewidth=5, color='black')
    plt.plot([0, 0], [-2, 2], 'k--')
    plt.ylim(-4, 4)
    plt.xlim(-8, 8)
    plt.xlabel('U(y)')
    plt.ylabel('y')
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_function(lambda x: 0*x, (-2, 2), title='No Flow')
```

```python
p = 1.5
rho = 1
nu = 1
L = 2
plot_function(lambda y: (p/(rho*nu))*(y**2-L**2), (-2, 2), title='Plane Poiseuille Flow')
```

```python
V_2 = 4
plot_function(lambda y: (V_2/L)*y, (-2, 2), title='Plane Couette Flow')
```

```python
plot_function(lambda y: (p/(rho*nu))*(y**2-L**2) + (V_2/L)*y, (-2, 2), title='Plane Poiseuille + Couette Flow')
```

```python
V_1 = 3 
p = 4
plot_function(
    lambda y: (p / (V_1 * rho)) * (
        (L / np.sinh(V_1 * L / nu)) * (np.exp(V_1 * y / nu) - np.cosh(V_1 * L / nu)) - y
    ),
    (-2, 2),
    title='Asymptotic Suction Profile (non-zero pressure gradient)'
)
```

```python
V_1 = -2
plot_function(
    lambda y: (V_2 / (2 * np.sinh(V_1 * L / nu))) * (np.exp(V_1 * y / nu) - np.exp(-V_1 * L / nu)),
    (-2, 2),
    title='Asymptotic Suction Profile (zero pressure gradient & V_1 < 0)'
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_unbounded_function(func, y_range, title):
    y = np.linspace(y_range[0], y_range[1], 400)
    x = func(y)
    plt.plot(x, y)
    y_arrows = np.linspace(y_range[0], y_range[1], 9)  # Adjust number of arrows
    for y_val in y_arrows:
        x_val = func(y_val)
        plt.annotate('', xy=(x_val, y_val), xytext=(0, y_val),
                     arrowprops=dict(arrowstyle="->", color='red'))
    plt.plot([-8, 8], [0, 0], linewidth=5, color='black')
    plt.plot([0, 0], [-2, 2], 'k--')
    plt.xlim(-8, 8)
    plt.xlabel('U(y)')
    plt.ylabel('y')
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
U_inf = 4
plot_unbounded_function(
    lambda y: U_inf * (1 - np.exp(V_1 * y / nu)),
    (0, 2),
    title='Unbounded Asymptotic Suction Profile (zero pressure gradient & V_1 < 0)'
)
```
